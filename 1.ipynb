Using the set_style() function of Seaborn we can set any of the themes 
available on Seaborn library. 
 Darkgrid 
 Whitegrid 
 Dark 
 White 
 Ticks

```python
plt.scatter(data.speeding,data.alcohol)
sns.set_style("dark")
plt.show()
```

We can remove the top and right axis spines using the despine() function.

```python
plt.scatter(data.speeding,data.alcohol)
sns.set_style("ticks")
# sns.set_style("dark")
sns.despine()
plt.show()
```

```python
sns.barplot(x='embark_town',y='age',data=dt)
```

Count plot 
The count plot can be thought of as a histogram across a categorical variable. 

```python
sns.set_context('paper') 
  
# load dataset 
dt = sns.load_dataset('titanic') 
dt
#create plot
sns.countplot(x='class',hue='who',data=dt,palette = 'magma')
plt.title('Survivors') 
plt.show() 
```

Point Plot 

```python
# loading dataset  
dt = sns.load_dataset("tips")  

sns.pointplot(x='day',y='tip',data=dt,hue='smoker')

plt.show()
```

Joint Plot 

```python
sns.set_style("dark") 
dt=sns.load_dataset('tips') 

sns.jointplot(x='total_bill',y='tip',data=dt,kind='reg',ci=None)
```

Regplot

```python
dt = sns.load_dataset("tips") 
ax = sns.regplot(x="total_bill", y="tip", data=dt,ci=None)
```
